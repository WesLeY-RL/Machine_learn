SCALLING


---
Min-Max Scaling √© uma t√©cnica de normaliza√ß√£o usada no pr√©-processamento de dados em aprendizado de m√°quina. Seu objetivo √© reescalar os valores de uma ou mais vari√°veis para um intervalo definido, geralmente entre 0 e 1. Isso √© feito para garantir que todas as features contribuam igualmente para o modelo, evitando que vari√°veis com escalas maiores dominem a aprendizagem.

A f√≥rmula utilizada √©:

Xescalado = ( X - Xmin) / ( Xmax - Xmin)

Essa transforma√ß√£o torna os dados mais adequados para algoritmos de machine learning sens√≠veis √† escala, como redes neurais, SVM e k-NN. Contudo, ela pode ser afetada por outliers, j√° que depende diretamente dos valores extremos das vari√°veis.


---
Em dados com outliers extremos, talvez seja melhor usar StandardScaler (padroniza√ß√£o com m√©dia 0 e desvio padr√£o 1).
Em machine learning, √© comum que as vari√°veis tenham escalas muito diferentes.


```python
# Min max scaling
data_scaler_minmax = preprocessing.MinMaxScaler(feature_range=(0, 1))
data_scaled_minmax = data_scaler_minmax.fit_transform(df)
print("\nMin max scaled data:\n", data_scaled_minmax)
```

Normalization (Normaliza√ß√£o)

---

Normaliza√ß√£o √© um processo usado para ajustar os valores das vari√°veis em um vetor de caracter√≠sticas para que fiquem em uma escala comum, facilitando compara√ß√µes e melhorando o desempenho de algoritmos de aprendizado de m√°quina.

As duas formas principais de normaliza√ß√£o s√£o:

L1 Normalization (Desvios Absolutos M√≠nimos)
Garante que a soma dos valores absolutos de cada linha seja igual a 1.

Resistente a outliers, pois n√£o d√° muito peso a valores extremos.

√ötil quando queremos reduzir o impacto de dados extremos nos c√°lculos.

```python
data_normalized_l1 = preprocessing.normalize(df, norm='l1')
print(data_normalized_l1)
```

Simplicidade geom√©trica: No espa√ßo vetorial, os vetores L1-normalizados ficam sobre um hiperplano cuja soma absoluta √© 1 ‚Äî uma forma mais ‚Äúpontuda‚Äù, boa pra t√©cnicas esparsas (como Lasso regression). Sim, √© isso. Pegue todos os valores de uma linha, aplique o valor absoluto (porque estamos lidando com dist√¢ncias), some tudo, e ent√£o divida cada elemento por essa soma.

L2 Normalization (M√≠nimos Quadrados)
Garante que a soma dos quadrados dos valores de cada linha seja igual a 1.

D√° mais peso a valores maiores (ou outliers), sendo mais sens√≠vel a eles.

Indicado quando os outliers s√£o importantes para a tarefa do modelo.

```python
data_normalized_l2 = preprocessing.normalize(df, norm='l2')
print(data_normalized_l2)
```

Logistic Regression Classifier


---


A regress√£o log√≠stica √© uma t√©cnica usada para estimar a rela√ß√£o entre vari√°veis de entrada (independentes) e vari√°veis de sa√≠da (dependentes), onde a sa√≠da √© categ√≥rica (ex: sim/n√£o, 0/1, classes).
Utiliza uma fun√ß√£o log√≠stica (ou sigmoide) para gerar probabilidades como sa√≠da.

A curva sigmoide transforma os valores de entrada em valores entre 0 e 1, o que permite interpretar a sa√≠da como uma probabilidade de pertencimento a uma classe.

Embora o nome sugira "regress√£o", ela √© usada para classifica√ß√£o.

√â uma t√©cnica simples e muito utilizada em machine learning, especialmente para problemas de classifica√ß√£o bin√°ria.

Baseia-se em m√©todos estat√≠sticos semelhantes √† regress√£o linear, mas com foco em minimizar o erro de classifica√ß√£o, e n√£o em prever valores cont√≠nuos.



```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Define sample input data
X = np.array([[3.1, 7.2], [4, 6.7], [2.9, 8], [5.1, 4.5], [6, 5], [5.6, 5],
[3.3, 0.4], [3.9, 0.9], [2.8, 1], [0.5, 3.4], [1, 4], [0.6, 4.9]])
y = np.array([0, 0, 0, 1, 1, 1, 2, 2, 2, 3, 3, 3])
```

```python
# Create the logistic regression classifier
classifier = linear_model.LogisticRegression(solver='liblinear', C=1)
```

```python
# Train the classifier
a = classifier.fit(X, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Fun√ß√£o para visualiza√ß√£o

def visualize_classifier(model, X, y, title="Visualiza√ß√£o do Classificador"):
    # Gera malha de pontos para desenhar a regi√£o de decis√£o
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plota a regi√£o de decis√£o
    plt.figure(figsize=(10, 8))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)

    # Plota os pontos de dados
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='green', s=100)

    # R√≥tulos e legenda
    plt.title(title, fontsize=16)
    plt.xlabel("Feature 1", fontsize=12)
    plt.ylabel("Feature 2", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.5)

    # Legenda personalizada
    classes = np.unique(y)
    legend_labels = [f"Classe {i}" for i in classes]
    handles = [plt.Line2D([0], [0], marker='o', color='black', label=label,
                          markerfacecolor=plt.cm.coolwarm(i / len(classes)), markersize=10, markeredgecolor='k')
               for i, label in enumerate(legend_labels)]
    plt.legend(handles=handles, title='Classes', loc='upper right')

    plt.tight_layout()
    plt.show()

# 5. Executa visualiza√ß√£o
visualize_classifier(classifier, X, y)

```

```python
visualize_classifier(a, X, y)
```

```python
# DIRETAMENTE DO LIVRO

def visualize_classifier(classifier, X, y):
  # Define the minimum and maximum values for X and Y
  # that will be used in the mesh grid
  min_x, max_x = X[:, 0].min() - 1.0, X[:, 0].max() + 1.0
  min_y, max_y = X[:, 1].min() - 1.0, X[:, 1].max() + 1.0
  # Define the step size to use in plotting the mesh grid
  mesh_step_size = 0.01
  # Define the mesh grid of X and Y values
  x_vals, y_vals = np.meshgrid(np.arange(min_x, max_x, mesh_step_size),
  np.arange(min_y, max_y, mesh_step_size))
  # Run the classifier on the mesh grid
  output = classifier.predict(np.c_[x_vals.ravel(), y_vals.ravel()])
  # Reshape the output array
  output = output.reshape(x_vals.shape)
  # Create a plot
  plt.figure()
  # Choose a color scheme for the plot
  plt.pcolormesh(x_vals, y_vals, output, cmap=plt.cm.gray)
  # Overlay the training points on the plot
  plt.scatter(X[:, 0], X[:, 1], c=y, s=75, edgecolors='black',
  linewidth=1, cmap=plt.cm.Paired)
  # Specify the boundaries of the plot
  plt.xlim(x_vals.min(), x_vals.max())
  plt.ylim(y_vals.min(), y_vals.max())
  # Specify the ticks on the X and Y axes
  plt.xticks((np.arange(int(X[:, 0].min() - 1), int(X[:, 0].max() + 1),
  1.0)))
  plt.yticks((np.arange(int(X[:, 1].min() - 1), int(X[:, 1].max() + 1),
  1.0)))
  plt.show()
```

```python
visualize_classifier(a, X, y)
```

Na√Øve Bayes Classifier


---


Resumo: O classificador Na√Øve Bayes √© um algoritmo de classifica√ß√£o baseado no Teorema de Bayes, que estima a probabilidade de uma classe com base em valores de atributos (features) de entrada. Ele assume que todas as features s√£o independentes entre si, mesmo que isso nem sempre seja verdade ‚Äî por isso o nome "na√Øve" (ing√™nuo).



O Teorema de Bayes calcula a probabilidade de um evento ocorrer com base em informa√ß√µes pr√©vias.

No Na√Øve Bayes, cada inst√¢ncia (linha de dados) √© representada como um vetor de caracter√≠sticas (features).

O algoritmo assume independ√™ncia total entre os atributos. Ou seja, ele considera que o valor de uma feature n√£o influencia o valor de outra.

Apesar da simplicidade, o Na√Øve Bayes √© eficiente e muito utilizado em textos, e-mails (como filtros de spam), e outras aplica√ß√µes.

```python
# 1. Importando bibliotecas necess√°rias
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# 2. Carregando o dataset Iris
iris = load_iris()

# 3. Separando os dados (X = caracter√≠sticas, y = r√≥tulos das classes)
X = iris.data  # array com 4 caracter√≠sticas por flor
y = iris.target  # 0 = setosa, 1 = versicolor, 2 = virginica

# 4. Dividindo em dados de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Criando e treinando o classificador Na√Øve Bayes
model = GaussianNB()  # usando o Na√Øve Bayes Gaussiano
model.fit(X_train, y_train)  # treinando o modelo

# 6. Fazendo previs√µes
y_pred = model.predict(X_test)

# 7. Avaliando o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Acur√°cia: {accuracy:.2f}")

# 8. Exemplo de previs√£o com novos dados
nova_flor = [[5.0, 3.4, 1.6, 0.4]]  # dados de uma flor n√£o rotulada
predicao = model.predict(nova_flor)
print(f"Classe prevista para a nova flor: {iris.target_names[predicao[0]]}")
```

1. Importa√ß√µes
load_iris: carrega o dataset Iris.

train_test_split: divide os dados em treino e teste.

GaussianNB: classificador Na√Øve Bayes que assume distribui√ß√£o gaussiana (normal) dos dados cont√≠nuos.

accuracy_score: mede a taxa de acerto do modelo.

2‚Äì3. Carregando e separando os dados
X: cont√©m as caracter√≠sticas (features) das flores.

y: cont√©m os r√≥tulos das classes (setosa, versicolor, virginica).

4. Divis√£o treino/teste
70% dos dados v√£o para treino, 30% para teste.

random_state=42 garante que o sorteio seja o mesmo em toda execu√ß√£o (reprodutibilidade).

5. Cria√ß√£o e treinamento
GaussianNB() √© usado pois as features s√£o cont√≠nuas (flores medidas em cm).

fit() ensina o modelo a "aprender" padr√µes dos dados de treino.

6. Previs√£o
predict() aplica o modelo aos dados de teste para prever as classes.

7. Avalia√ß√£o
A acur√°cia mostra qual porcentagem das flores de teste foram corretamente classificadas.

8. Previs√£o real
Inserimos uma flor com medidas conhecidas mas sem classe.

O modelo retorna a classe que ele acha mais prov√°vel, com base no treinamento.

Confusion Matrix


---
√â uma tabela usada para avaliar o desempenho de um modelo de classifica√ß√£o. Ela mostra como as previs√µes do modelo se comparam com os valores reais (tamb√©m chamados de "ground truth").
Para cada classe, o modelo pode:

Acertar a previs√£o

Errar a previs√£o (e confundir com outra classe)

A matriz de confus√£o organiza essas informa√ß√µes de forma que voc√™ veja quantos acertos e erros o modelo cometeu em cada categoria.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Define sample labels
true_labels = [2, 0, 0, 2, 4, 4, 1, 0, 3, 3, 3]
pred_labels = [2, 1, 0, 2, 4, 3, 1, 0, 1, 3, 3]
```

```python
# Create confusion matrix
confusion_mat = confusion_matrix(true_labels, pred_labels)
```

```python
# Visualize confusion matrix
plt.imshow(confusion_mat, interpolation='nearest', cmap=plt.cm.gray)
plt.title('Confusion matrix')
plt.colorbar()
ticks = np.arange(5)
plt.xticks(ticks, ticks)
plt.yticks(ticks, ticks)
plt.ylabel('True labels')
plt.xlabel('Predicted labels')
plt.show()
```

```python
# Classification report
targets = ['Class-0', 'Class-1', 'Class-2', 'Class-3', 'Class-4']
print('\n', classification_report(true_labels, pred_labels,
target_names=targets))
```

Support Vector Machine

---
O SVM funciona tentando separar as classes de dados com uma fronteira de decis√£o ideal chamada:

üîµ Hiperplano (hyperplane)
√â uma linha que separa os dados no caso 2D.

Em dimens√µes maiores, essa "linha" vira um plano (3D) ou um hiperplano (N-dimensional).

O objetivo √© encontrar o hiperplano que separa melhor os dados ‚Äî ou seja, que maximiza a margem entre as duas classes.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# 1. Gerar dados sint√©ticos (2 classes)
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)

# 2. Criar o modelo SVM
model = SVC(kernel='linear')  # SVM com kernel linear (linha como hiperplano)
model.fit(X, y)

# 3. Visualizar os dados e o hiperplano
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=30)

# 4. Plotar o hiperplano
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Criar grid para plotar
xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = model.decision_function(xy).reshape(XX.shape)

# Desenhar as margens e o hiperplano
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], linestyles=['--', '-', '--'])

# 5. Destacar os vetores de suporte
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=100, linewidth=1, facecolors='none', edgecolors='k')

plt.title('SVM com hiperplano e vetores de suporte')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

Regress√£o

---

Regress√£o √© um tipo de problema de aprendizado de m√°quina onde o objetivo √© prever um valor cont√≠nuo (por exemplo, sal√°rio, pre√ßo de uma casa, temperatura, etc.), com base em uma ou mais vari√°veis de entrada.

1. Previs√£o de valores cont√≠nuos
A sa√≠da (vari√°vel dependente) √© um n√∫mero real cont√≠nuo, como:

Altura

Peso

Lucro

Temperatura

Isso √© diferente de classifica√ß√£o, onde a sa√≠da √© uma categoria (ex: ‚Äúgato‚Äù ou ‚Äúcachorro‚Äù, ou 0 e 1).

Vari√°veis de entrada = independentes (tamb√©m chamadas de "preditores" ou "features").

Vari√°vel de sa√≠da = dependente (depende das entradas).
Estimar a rela√ß√£o matem√°tica entre entrada e sa√≠da.

Queremos saber como a mudan√ßa em uma ou mais vari√°veis de entrada afeta o valor da sa√≠da.

Exemplo simples:

‚ÄúQuanto mais anos de experi√™ncia uma pessoa tem, maior tende a ser o sal√°rio.‚Äù
Linear Regression
Sup√µe que a rela√ß√£o entre entrada e sa√≠da √© linear.

Simples e eficiente.

Exemplo: sal√°rio = a * anos_de_experi√™ncia + b

 Correla√ß√£o entre vari√°veis
Nem sempre as vari√°veis de entrada s√£o independentes entre si.

Em muitos casos reais, as features se correlacionam (ex: peso e altura).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Metragem das casas (em metros quadrados)
X = np.array([[50], [60], [70], [80], [90], [100], [110], [120]])

# Pre√ßo correspondente (em mil R$)
y = np.array([150, 180, 200, 210, 240, 250, 270, 300])
```

```python
# Criar o modelo
modelo = LinearRegression()

# Treinar o modelo
modelo.fit(X, y)
```

```python
# Previs√£o do modelo
y_pred = modelo.predict(X)

# Plotar os pontos reais
plt.scatter(X, y, color='blue', label='Dados reais')

# Plotar a reta de regress√£o
plt.plot(X, y_pred, color='red', linewidth=2, label='Regress√£o linear')

# T√≠tulos
plt.xlabel('Metragem (m¬≤)')
plt.ylabel('Pre√ßo (mil R$)')
plt.title('Regress√£o Linear: Pre√ßo x Metragem')
plt.legend()
plt.grid(True)
plt.show()
```

Construindo um regressor multivari√°vel (Multivariable Regressor)

---


üîç O que significa?
At√© agora, a regress√£o foi feita com uma √∫nica vari√°vel independente (ex: prever o pre√ßo de uma casa com base apenas na metragem). Isso se chama regress√£o univariada.

Agora, vamos usar mais de uma vari√°vel de entrada (input) para prever a sa√≠da. Isso √© chamado de:

‚û§ Regress√£o multivariada (ou m√∫ltipla)

Exemplo real
Suponha que voc√™ quer prever o pre√ßo de uma casa com base em:

Metragem quadrada

N√∫mero de quartos

Idade do im√≥vel

Nesse caso, o modelo precisa analisar o efeito combinado de v√°rias vari√°veis para estimar o resultado (pre√ßo).



```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Cada linha representa uma casa: [metragem, quartos, idade]
X = np.array([
    [50, 1, 15],
    [60, 2, 10],
    [70, 2, 8],
    [80, 3, 5],
    [90, 3, 3],
    [100, 4, 2],
    [110, 4, 1],
    [120, 5, 1]
])

# Pre√ßos em mil reais
y = np.array([180, 200, 220, 250, 270, 300, 320, 350])
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

Esses coeficientes indicam quanto o pre√ßo aumenta para cada:

1 m¬≤ a mais

1 quarto a mais

1 ano a menos de idade (quanto mais nova, maior o valor)

```python
print("Coeficientes (pesos):", modelo.coef_)
print("Intercepto (b):", modelo.intercept_)
```

```python
nova_casa = np.array([[85, 3, 5]])
previsao = modelo.predict(nova_casa)

print(f"Pre√ßo estimado: R$ {previsao[0]:.2f} mil")
```
