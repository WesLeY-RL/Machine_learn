MEAN REMOVAL
---
Mean Removal é uma técnica de pré-processamento usada em aprendizado de máquina e análise de dados, onde a média dos dados é subtraída de cada valor da amostra, para que os dados fiquem com média zero.

---

Centraliza os dados em torno de zero, o que ajuda muitos algoritmos (ex: SVM, PCA) a convergirem mais rápido.

Remove o viés dos dados — por exemplo, se uma variável está sistematicamente deslocada, ela pode distorcer o modelo.





```python
# Print mean and standard deviation
print("\nBEFORE:")
print("Mean ou Média =", df.mean(axis=0))
print("Std deviation  ou Desvio Padrão =", df.std(axis=0))
```

```python
# Remove mean
data_scaled = preprocessing.scale(df)
print("\nAFTER:")
print("Mean =", data_scaled.mean(axis=0))
print("Std deviation =", data_scaled.std(axis=0))
```

SCALLING


---
Min-Max Scaling é uma técnica de normalização usada no pré-processamento de dados em aprendizado de máquina. Seu objetivo é reescalar os valores de uma ou mais variáveis para um intervalo definido, geralmente entre 0 e 1. Isso é feito para garantir que todas as features contribuam igualmente para o modelo, evitando que variáveis com escalas maiores dominem a aprendizagem.

A fórmula utilizada é:

Xescalado = ( X - Xmin) / ( Xmax - Xmin)

Essa transformação torna os dados mais adequados para algoritmos de machine learning sensíveis à escala, como redes neurais, SVM e k-NN. Contudo, ela pode ser afetada por outliers, já que depende diretamente dos valores extremos das variáveis.


---
Em dados com outliers extremos, talvez seja melhor usar StandardScaler (padronização com média 0 e desvio padrão 1).
Em machine learning, é comum que as variáveis tenham escalas muito diferentes.


```python
# Min max scaling
data_scaler_minmax = preprocessing.MinMaxScaler(feature_range=(0, 1))
data_scaled_minmax = data_scaler_minmax.fit_transform(df)
print("\nMin max scaled data:\n", data_scaled_minmax)
```

```python
# Min max scaling
data_scaler_minmax = preprocessing.MinMaxScaler(feature_range=(0, 1))
data_scaled_minmax = data_scaler_minmax.fit_transform(df)
print("\nMin max scaled data:\n", data_scaled_minmax)
```

StandardScaler – Padronização (Média 0 e Desvio Padrão 1)


---
StandardScaler é uma técnica de padronização (standardization) usada para pré-processar dados antes de aplicar algoritmos de aprendizado de máquina. Ao contrário da normalização (como o Min-Max Scaling), a padronização transforma os dados para que tenham:

Média igual a 0

Desvio padrão igual a 1

Xpadronizado = (X - 𝜇) / σ

X é o valor original,


μ é a média dos dados,


σ é o desvio padrão.


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dados_padronizados = scaler.fit_transform(df)
print(dados_padronizados)
```

Normalization (Normalização)

---

Normalização é um processo usado para ajustar os valores das variáveis em um vetor de características para que fiquem em uma escala comum, facilitando comparações e melhorando o desempenho de algoritmos de aprendizado de máquina.

As duas formas principais de normalização são:

L1 Normalization (Desvios Absolutos Mínimos)
Garante que a soma dos valores absolutos de cada linha seja igual a 1.

Resistente a outliers, pois não dá muito peso a valores extremos.

Útil quando queremos reduzir o impacto de dados extremos nos cálculos.

```python
data_normalized_l1 = preprocessing.normalize(df, norm='l1')
print(data_normalized_l1)
```

Simplicidade geométrica: No espaço vetorial, os vetores L1-normalizados ficam sobre um hiperplano cuja soma absoluta é 1 — uma forma mais “pontuda”, boa pra técnicas esparsas (como Lasso regression). Sim, é isso. Pegue todos os valores de uma linha, aplique o valor absoluto (porque estamos lidando com distâncias), some tudo, e então divida cada elemento por essa soma.

L2 Normalization (Mínimos Quadrados)
Garante que a soma dos quadrados dos valores de cada linha seja igual a 1.

Dá mais peso a valores maiores (ou outliers), sendo mais sensível a eles.

Indicado quando os outliers são importantes para a tarefa do modelo.

```python
data_normalized_l2 = preprocessing.normalize(df, norm='l2')
print(data_normalized_l2)
```

Logistic Regression Classifier


---


A regressão logística é uma técnica usada para estimar a relação entre variáveis de entrada (independentes) e variáveis de saída (dependentes), onde a saída é categórica (ex: sim/não, 0/1, classes).
Utiliza uma função logística (ou sigmoide) para gerar probabilidades como saída.

A curva sigmoide transforma os valores de entrada em valores entre 0 e 1, o que permite interpretar a saída como uma probabilidade de pertencimento a uma classe.

Embora o nome sugira "regressão", ela é usada para classificação.

É uma técnica simples e muito utilizada em machine learning, especialmente para problemas de classificação binária.

Baseia-se em métodos estatísticos semelhantes à regressão linear, mas com foco em minimizar o erro de classificação, e não em prever valores contínuos.



```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Define sample input data
X = np.array([[3.1, 7.2], [4, 6.7], [2.9, 8], [5.1, 4.5], [6, 5], [5.6, 5],
[3.3, 0.4], [3.9, 0.9], [2.8, 1], [0.5, 3.4], [1, 4], [0.6, 4.9]])
y = np.array([0, 0, 0, 1, 1, 1, 2, 2, 2, 3, 3, 3])
```

```python
# Create the logistic regression classifier
classifier = linear_model.LogisticRegression(solver='liblinear', C=1)
```

```python
# Train the classifier
a = classifier.fit(X, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Função para visualização

def visualize_classifier(model, X, y, title="Visualização do Classificador"):
    # Gera malha de pontos para desenhar a região de decisão
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plota a região de decisão
    plt.figure(figsize=(10, 8))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)

    # Plota os pontos de dados
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='green', s=100)

    # Rótulos e legenda
    plt.title(title, fontsize=16)
    plt.xlabel("Feature 1", fontsize=12)
    plt.ylabel("Feature 2", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.5)

    # Legenda personalizada
    classes = np.unique(y)
    legend_labels = [f"Classe {i}" for i in classes]
    handles = [plt.Line2D([0], [0], marker='o', color='black', label=label,
                          markerfacecolor=plt.cm.coolwarm(i / len(classes)), markersize=10, markeredgecolor='k')
               for i, label in enumerate(legend_labels)]
    plt.legend(handles=handles, title='Classes', loc='upper right')

    plt.tight_layout()
    plt.show()

# 5. Executa visualização
visualize_classifier(classifier, X, y)

```

```python
visualize_classifier(a, X, y)
```

```python
# DIRETAMENTE DO LIVRO

def visualize_classifier(classifier, X, y):
  # Define the minimum and maximum values for X and Y
  # that will be used in the mesh grid
  min_x, max_x = X[:, 0].min() - 1.0, X[:, 0].max() + 1.0
  min_y, max_y = X[:, 1].min() - 1.0, X[:, 1].max() + 1.0
  # Define the step size to use in plotting the mesh grid
  mesh_step_size = 0.01
  # Define the mesh grid of X and Y values
  x_vals, y_vals = np.meshgrid(np.arange(min_x, max_x, mesh_step_size),
  np.arange(min_y, max_y, mesh_step_size))
  # Run the classifier on the mesh grid
  output = classifier.predict(np.c_[x_vals.ravel(), y_vals.ravel()])
  # Reshape the output array
  output = output.reshape(x_vals.shape)
  # Create a plot
  plt.figure()
  # Choose a color scheme for the plot
  plt.pcolormesh(x_vals, y_vals, output, cmap=plt.cm.gray)
  # Overlay the training points on the plot
  plt.scatter(X[:, 0], X[:, 1], c=y, s=75, edgecolors='black',
  linewidth=1, cmap=plt.cm.Paired)
  # Specify the boundaries of the plot
  plt.xlim(x_vals.min(), x_vals.max())
  plt.ylim(y_vals.min(), y_vals.max())
  # Specify the ticks on the X and Y axes
  plt.xticks((np.arange(int(X[:, 0].min() - 1), int(X[:, 0].max() + 1),
  1.0)))
  plt.yticks((np.arange(int(X[:, 1].min() - 1), int(X[:, 1].max() + 1),
  1.0)))
  plt.show()
```

```python
visualize_classifier(a, X, y)
```

Naïve Bayes Classifier


---


Resumo: O classificador Naïve Bayes é um algoritmo de classificação baseado no Teorema de Bayes, que estima a probabilidade de uma classe com base em valores de atributos (features) de entrada. Ele assume que todas as features são independentes entre si, mesmo que isso nem sempre seja verdade — por isso o nome "naïve" (ingênuo).



O Teorema de Bayes calcula a probabilidade de um evento ocorrer com base em informações prévias.

No Naïve Bayes, cada instância (linha de dados) é representada como um vetor de características (features).

O algoritmo assume independência total entre os atributos. Ou seja, ele considera que o valor de uma feature não influencia o valor de outra.

Apesar da simplicidade, o Naïve Bayes é eficiente e muito utilizado em textos, e-mails (como filtros de spam), e outras aplicações.

```python
# 1. Importando bibliotecas necessárias
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# 2. Carregando o dataset Iris
iris = load_iris()

# 3. Separando os dados (X = características, y = rótulos das classes)
X = iris.data  # array com 4 características por flor
y = iris.target  # 0 = setosa, 1 = versicolor, 2 = virginica

# 4. Dividindo em dados de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Criando e treinando o classificador Naïve Bayes
model = GaussianNB()  # usando o Naïve Bayes Gaussiano
model.fit(X_train, y_train)  # treinando o modelo

# 6. Fazendo previsões
y_pred = model.predict(X_test)

# 7. Avaliando o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia: {accuracy:.2f}")

# 8. Exemplo de previsão com novos dados
nova_flor = [[5.0, 3.4, 1.6, 0.4]]  # dados de uma flor não rotulada
predicao = model.predict(nova_flor)
print(f"Classe prevista para a nova flor: {iris.target_names[predicao[0]]}")
```

1. Importações
load_iris: carrega o dataset Iris.

train_test_split: divide os dados em treino e teste.

GaussianNB: classificador Naïve Bayes que assume distribuição gaussiana (normal) dos dados contínuos.

accuracy_score: mede a taxa de acerto do modelo.

2–3. Carregando e separando os dados
X: contém as características (features) das flores.

y: contém os rótulos das classes (setosa, versicolor, virginica).

4. Divisão treino/teste
70% dos dados vão para treino, 30% para teste.

random_state=42 garante que o sorteio seja o mesmo em toda execução (reprodutibilidade).

5. Criação e treinamento
GaussianNB() é usado pois as features são contínuas (flores medidas em cm).

fit() ensina o modelo a "aprender" padrões dos dados de treino.

6. Previsão
predict() aplica o modelo aos dados de teste para prever as classes.

7. Avaliação
A acurácia mostra qual porcentagem das flores de teste foram corretamente classificadas.

8. Previsão real
Inserimos uma flor com medidas conhecidas mas sem classe.

O modelo retorna a classe que ele acha mais provável, com base no treinamento.

Confusion Matrix


---
É uma tabela usada para avaliar o desempenho de um modelo de classificação. Ela mostra como as previsões do modelo se comparam com os valores reais (também chamados de "ground truth").
Para cada classe, o modelo pode:

Acertar a previsão

Errar a previsão (e confundir com outra classe)

A matriz de confusão organiza essas informações de forma que você veja quantos acertos e erros o modelo cometeu em cada categoria.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Define sample labels
true_labels = [2, 0, 0, 2, 4, 4, 1, 0, 3, 3, 3]
pred_labels = [2, 1, 0, 2, 4, 3, 1, 0, 1, 3, 3]
```

```python
# Create confusion matrix
confusion_mat = confusion_matrix(true_labels, pred_labels)
```

```python
# Visualize confusion matrix
plt.imshow(confusion_mat, interpolation='nearest', cmap=plt.cm.gray)
plt.title('Confusion matrix')
plt.colorbar()
ticks = np.arange(5)
plt.xticks(ticks, ticks)
plt.yticks(ticks, ticks)
plt.ylabel('True labels')
plt.xlabel('Predicted labels')
plt.show()
```

```python
# Classification report
targets = ['Class-0', 'Class-1', 'Class-2', 'Class-3', 'Class-4']
print('\n', classification_report(true_labels, pred_labels,
target_names=targets))
```

Support Vector Machine

---
O SVM funciona tentando separar as classes de dados com uma fronteira de decisão ideal chamada:

🔵 Hiperplano (hyperplane)
É uma linha que separa os dados no caso 2D.

Em dimensões maiores, essa "linha" vira um plano (3D) ou um hiperplano (N-dimensional).

O objetivo é encontrar o hiperplano que separa melhor os dados — ou seja, que maximiza a margem entre as duas classes.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# 1. Gerar dados sintéticos (2 classes)
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)

# 2. Criar o modelo SVM
model = SVC(kernel='linear')  # SVM com kernel linear (linha como hiperplano)
model.fit(X, y)

# 3. Visualizar os dados e o hiperplano
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=30)

# 4. Plotar o hiperplano
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Criar grid para plotar
xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = model.decision_function(xy).reshape(XX.shape)

# Desenhar as margens e o hiperplano
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], linestyles=['--', '-', '--'])

# 5. Destacar os vetores de suporte
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=100, linewidth=1, facecolors='none', edgecolors='k')

plt.title('SVM com hiperplano e vetores de suporte')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

Regressão

---

Regressão é um tipo de problema de aprendizado de máquina onde o objetivo é prever um valor contínuo (por exemplo, salário, preço de uma casa, temperatura, etc.), com base em uma ou mais variáveis de entrada.

1. Previsão de valores contínuos
A saída (variável dependente) é um número real contínuo, como:

Altura

Peso

Lucro

Temperatura

Isso é diferente de classificação, onde a saída é uma categoria (ex: “gato” ou “cachorro”, ou 0 e 1).

Variáveis de entrada = independentes (também chamadas de "preditores" ou "features").

Variável de saída = dependente (depende das entradas).
Estimar a relação matemática entre entrada e saída.

Queremos saber como a mudança em uma ou mais variáveis de entrada afeta o valor da saída.

Exemplo simples:

“Quanto mais anos de experiência uma pessoa tem, maior tende a ser o salário.”
Linear Regression
Supõe que a relação entre entrada e saída é linear.

Simples e eficiente.

Exemplo: salário = a * anos_de_experiência + b

 Correlação entre variáveis
Nem sempre as variáveis de entrada são independentes entre si.

Em muitos casos reais, as features se correlacionam (ex: peso e altura).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Metragem das casas (em metros quadrados)
X = np.array([[50], [60], [70], [80], [90], [100], [110], [120]])

# Preço correspondente (em mil R$)
y = np.array([150, 180, 200, 210, 240, 250, 270, 300])
```

```python
# Criar o modelo
modelo = LinearRegression()

# Treinar o modelo
modelo.fit(X, y)
```

```python
# Previsão do modelo
y_pred = modelo.predict(X)

# Plotar os pontos reais
plt.scatter(X, y, color='blue', label='Dados reais')

# Plotar a reta de regressão
plt.plot(X, y_pred, color='red', linewidth=2, label='Regressão linear')

# Títulos
plt.xlabel('Metragem (m²)')
plt.ylabel('Preço (mil R$)')
plt.title('Regressão Linear: Preço x Metragem')
plt.legend()
plt.grid(True)
plt.show()
```

Construindo um regressor multivariável (Multivariable Regressor)

---


🔍 O que significa?
Até agora, a regressão foi feita com uma única variável independente (ex: prever o preço de uma casa com base apenas na metragem). Isso se chama regressão univariada.

Agora, vamos usar mais de uma variável de entrada (input) para prever a saída. Isso é chamado de:

➤ Regressão multivariada (ou múltipla)

Exemplo real
Suponha que você quer prever o preço de uma casa com base em:

Metragem quadrada

Número de quartos

Idade do imóvel

Nesse caso, o modelo precisa analisar o efeito combinado de várias variáveis para estimar o resultado (preço).



```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Cada linha representa uma casa: [metragem, quartos, idade]
X = np.array([
    [50, 1, 15],
    [60, 2, 10],
    [70, 2, 8],
    [80, 3, 5],
    [90, 3, 3],
    [100, 4, 2],
    [110, 4, 1],
    [120, 5, 1]
])

# Preços em mil reais
y = np.array([180, 200, 220, 250, 270, 300, 320, 350])
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

Esses coeficientes indicam quanto o preço aumenta para cada:

1 m² a mais

1 quarto a mais

1 ano a menos de idade (quanto mais nova, maior o valor)

```python
print("Coeficientes (pesos):", modelo.coef_)
print("Intercepto (b):", modelo.intercept_)
```

```python
nova_casa = np.array([[85, 3, 5]])
previsao = modelo.predict(nova_casa)

print(f"Preço estimado: R$ {previsao[0]:.2f} mil")
```

```python
# --- Plot 3D (usando só duas variáveis para visualização: metragem e quartos)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Eixos
metragem = X[:, 0]
quartos = X[:, 1]

# Plot dos pontos reais
ax.scatter(metragem, quartos, y, color='blue', label='Dados reais')

# Criando a malha para previsão da superfície
m_grid, q_grid = np.meshgrid(
    np.linspace(min(metragem), max(metragem), 10),
    np.linspace(min(quartos), max(quartos), 10)
)
# Ajustar a idade como constante (ex: 5 anos) para simplificar visualização
idade_constante = 5
entrada_modelo = np.c_[m_grid.ravel(), q_grid.ravel(), idade_constante * np.ones(m_grid.size)]
preco_previsto = modelo.predict(entrada_modelo).reshape(m_grid.shape)

# Superfície de regressão
ax.plot_surface(m_grid, q_grid, preco_previsto, alpha=0.5, color='red', label='Plano de regressão')

# Rótulos
ax.set_xlabel('Metragem (m²)')
ax.set_ylabel('Quartos')
ax.set_zlabel('Preço (mil R$)')
ax.set_title('Regressão Linear Multivariada - Preço de Casa')

plt.show()
```
